## 📚 Import Libraries

```python
# ============================================
# Title: Import Essential Libraries
# Description: Import Pandas and NumPy for data manipulation, and Matplotlib/Seaborn for visualization.
# ============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 🛠️ Install Core Python Packages

```python
df = pd.read_csv('c:/Users/kayal/Downloads/customer_shopping_behavior.csv')
```

```python
# ============================================
# Display first 5 rows to understand the structure
# ============================================

df.head()
```

```python
# Display basic information about dataset (columns, data types, null values)

df.info()
```

```python
# ============================================
# Title: Summary Statistics of Dataset
# Description: Display descriptive statistics for all columns, including numeric and categorical features.
# ============================================

df.describe(include="all")
```

```python
# ============================================
# Title: Clean and Standardize Column Names
# Description: Convert column names to lowercase, replace spaces with underscores, 
#              and rename specific columns for consistency.
# ============================================

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
# ============================================
# Title: Impute Missing Review Ratings with Category Median
# Description: For each product category, fill missing 'review_rating' values 
#              with the median rating of that category to preserve distribution.
# ============================================

df['review_rating'] = df.groupby('category')['review_rating'] \
  .transform(lambda x: x.fillna(x.median()))
```

```python
# Check for missing values in each column

df.isnull().sum()
```

```python
# Numerical summary using NumPy
print("Mean Age:", np.mean(df['Age']))
print("Median Age:", np.median(df['Age']))
print("Standard Deviation:", np.std(df['Age']))
```

## 📊 Age Distribution

```python
# ============================================
# Title: Plot Age Distribution
# Description: Visualize the distribution of customer ages using a histogram with KDE.
# ============================================

plt.figure()
sns.histplot(df['Age'], kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

## 📦 Category vs Review Rating

```python
# ============================================
# Title: Boxplot of Category vs Review Rating
# Description: Visualize the distribution of review ratings across different product categories.
# ============================================

plt.figure(figsize=(10,5))
sns.boxplot(x='category', y='review_rating', data=df)
plt.xticks(rotation=45)
plt.title("Category vs Review Rating")
plt.xlabel("Category")
plt.ylabel("Review Rating")
plt.show()
```

## 🔗 Correlation Matrix

```python
# ============================================
# Title: Correlation Heatmap
# Description: Display the correlation between numerical features using a heatmap.
# ============================================

plt.figure(figsize=(8,6))
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

## ⏱️ Convert Purchase Frequency to Days

```python
# ============================================
# Title: Map Purchase Frequency to Days
# Description: Convert categorical purchase frequencies into numeric days for analysis.
# ============================================

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

## 👥 Age vs Purchase Frequency

```python
# ============================================
# Title: Scatter Plot of Age vs Purchase Frequency
# Description: Visualize the relationship between customer age and purchase frequency (in days).
# ============================================

plt.figure()
sns.scatterplot(x='age', y='purchase_frequency_days', data=df)
plt.title("Age vs Purchase Frequency")
plt.xlabel("Age")
plt.ylabel("Purchase Frequency (Days)")
plt.show()
```

## 👥 Create Age Groups

```python
# ============================================
# Title: Categorize Customers by Age Group
# Description: Divide customers into quartile-based age groups: Young Adult, Adult, Middle-aged, Senior.
# ============================================

labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels, duplicates='drop')
```

## 👥 Age Group Distribution

```python
# ============================================
# Title: Countplot of Age Groups
# Description: Visualize the number of customers in each age group category.
# ============================================

plt.figure()
sns.countplot(x='age_group', data=df)
plt.title("Age Group Distribution")
plt.xlabel("Age Group")
plt.ylabel("Number of Customers")
plt.show()
```

## ⭐ Average Rating per Category

```python
# ============================================
# Title: Average Review Rating by Category
# Description: Calculate the mean review rating for each product category and visualize it as a horizontal bar chart.
# ============================================

# Group by category and calculate mean review rating
grouped = df.groupby('category')['review_rating'].mean().sort_values()

# Plot horizontal bar chart
plt.figure()
grouped.plot(kind='barh')
plt.title("Average Rating per Category")
plt.xlabel("Average Review Rating")
plt.ylabel("Category")
plt.show()
```
